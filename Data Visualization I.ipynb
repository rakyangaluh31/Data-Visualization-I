```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Ritz_Jager_Data.csv', delimiter = ',')
df
```

Dataframe is still consisted of one column and it needs to be splitted.

### Splitting Dataframe

```python
list = df['hotel_type,is_canceled,lead_time,arrival_date_year,arrival_date_month,arrival_date_week_number,arrival_date_day_of_month,stays_in_weekend_nights,stays_in_week_nights,adults,children,babies,meal_type,country_origin,market_segment,distribution_channel,is_repeated_guest,previous_cancellations,previous_bookings_not_canceled,reserved_room_type,assigned_room_type,booking_changes,deposit_type,agent,company,days_in_waiting_list,customer_type,adr,required_car_parking_spaces,total_of_special_requests,reservation_status,reservation_status_date']
df_new = list.apply(lambda x: pd.Series(x.split(',')))
```

```python
df_new
```

```python
df_column = ['hotel_type', 'is_canceled', 'lead_time', 'arrival_date_year', 'arrival_date_month', 'arrival_date_week_number', 'arrival_date_day_of_month', 'stays_in_weekend_nights', 'stays_in_week_nights', 'adults', 'children', 'babies', 'meal_type', 'country_origin', 'market_segment', 'distribution_channel', 'is_repeated_guest', 'previous_cancellations', 'previous_bookings_not_canceled', 'reserved_room_type', 'assigned_room_type', 'booking_changes', 'deposit_type', 'agent', 'company', 'days_in_waiting_list', 'customer_type', 'adr', 'required_car_parking_spaces', 'total_of_special_requests', 'reservation_status', 'reservation_status_date']

#change all the columns name
df_new.columns = df_column
df_new
```

```python
df_new.info()
```

```python
df_new['stays_in_weekend_nights'] = df_new['stays_in_weekend_nights'].astype(float, errors = 'raise')
df_new['stays_in_week_nights'] = df_new['stays_in_week_nights'].astype(float, errors = 'raise')
```

#### 2. What month is the busiest (based on the highest number of bookings)?

```python
df_3 = pd.DataFrame(df_new.groupby(['arrival_date_year','arrival_date_month'])['is_canceled'].count())
df_3 = df_3.reset_index()
```

#### Graph in 2015

```python
#Graph in 2015
df_3_2015 = df_3.loc[df_3['arrival_date_year'] == '2015']

plt.bar(df_3_2015['arrival_date_month'],df_3_2015['is_canceled'])
plt.xlabel('Month')
plt.ylabel('Bookings')
plt.title('Bookings in 2015')
```

In 2015, September has the highest number of bookings.

#### Graph in 2016

```python
#Graph in 2016
df_3_2016 = df_3.loc[df_3['arrival_date_year'] == '2016']

plt.figure(figsize=(12,6))
plt.bar(df_3_2016['arrival_date_month'], df_3_2016['is_canceled'], width = 0.6)
plt.xlabel('Month')
plt.ylabel('Bookings')
plt.title('Bookings in 2016')
```

In 2016, Oktober has the highest number of bookings. 

#### Graph in 2017

```python
#Graph in 2017
df_3_2017 = df_3.loc[df_3['arrival_date_year'] == '2017']

plt.figure(figsize=(12,6))
plt.bar(df_3_2017['arrival_date_month'],df_3_2017['is_canceled'], width = 0.6)
plt.xlabel('Month')
plt.ylabel('Bookings')
plt.title('Bookings in 2017')
```

In 2017, May has the highest number of bookings. 

#### 3. How many nights the guests stay at the hotel in average ?

```python
df_night = df_new[['stays_in_weekend_nights','stays_in_week_nights']].mean()
df_night = pd.DataFrame(df_night).reset_index()
df_night = df_night.rename(columns ={'index':'stay at night', 0:'mean'})
df_night
```

```python
plt.bar(df_night['stay at night'], df_night['mean'])
plt.xlabel('Stay at Night')
plt.ylabel('Rerata Malam Menginap')
plt.title('Rerata Malam Tamu Menginap pada Tahun 2015 - 2017')
```

In the weekend, the guests stay around 1 day in average. In the weekdays, the guests stay around 2.5 days in average.

#### 4. How many bookings are made based on market segments ?

```python
df_4 = pd.DataFrame(df_new['market_segment'].value_counts())
df_4 = df_4.reset_index()
df_4['market segment'] = df_4['index']
df_4['count'] = df_4['market_segment']
df_4
```

```python
plt.figure(figsize=(12,6))
plt.bar(df_4['market segment'], df_4['count'])
plt.xlabel('Market Segment')
plt.ylabel('Bookings')
plt.title('Bookings based on Market Segment')
```

Based on the market segment, the bookings mostly come from Online TA with approximately 57k bookings. While Offline TA generate around 24k bookings. The rest come from groups, direct market, corporate, complementary, and aviation. 

#### 5. How many bookings are canceled per room type ?

```python
df_5 = pd.DataFrame(df_new.groupby(['reserved_room_type'])['is_canceled'].value_counts())
df_5 = df_5.rename(columns = {'is_canceled':'count'})
df_5 = df_5.reset_index()

df_5 = df_5.rename(columns = {'is_canceled':'status'})
df_5 = df_5.loc[df_5['status'] == '1']
df_5
```

```python
plt.figure(figsize=(12,6))
plt.bar(df_5['reserved_room_type'], df_5['count'])
plt.xlabel('Room Type')
plt.ylabel('Canceled Bookings')
plt.title('Canceled Bookings per Room Type in 2015 - 2017')
```

#### 6. What month with the highest number of cancelation ? 

```python
df_6 = df_new.groupby(['arrival_date_year','arrival_date_month'])['is_canceled'].value_counts().unstack().reset_index()
df_6 = df_6.rename(columns = {'1':'Canceled'})
df_6 = df_6.rename(columns = {'0':'Not Canceled'})
df_6
```

#### Graph in 2015

```python
#Graph in 2015
df_6_2015 = df_6.loc[df_6['arrival_date_year'] == '2015']

plt.figure(figsize=(10,5))
plt.bar(df_6_2015['arrival_date_month'],df_6_2015['Canceled'], width = 0.6)
plt.xlabel('Month')
plt.ylabel('Canceled Booking')
plt.title('Canceled Bookings in 2015')
```

In 2015, September has the highest number of canceled bookings with 2094 cancelation.

#### Graph in 2016

```python
#Grafik Tahun 2016
df_6_2016 = df_6.loc[df_6['arrival_date_year'] == '2016']

plt.figure(figsize=(11,8))
plt.bar(df_6_2016['arrival_date_month'],df_6_2016['Canceled'], width = 0.6)
plt.xlabel('Month')
plt.ylabel('Canceled Booking')
plt.title('Canceled Bookings in 2016')
```

In 2016, October has the highest Canceled Bookings with 2514 cancelation.

#### Graph in 2017

```python
#Grafik Tahun 2017
df_6_2017 = df_6.loc[df_6['arrival_date_year'] == '2017']

plt.figure(figsize=(10,6))
plt.bar(df_6_2017['arrival_date_month'],df_6_2017['Canceled'], width = 0.6)
plt.xlabel('Month')
plt.ylabel('Canceled Booking')
plt.title('Canceled Bookings in 2017')
```

In 2017, May has the highest Canceled Bookings with 2762 cancelation.
